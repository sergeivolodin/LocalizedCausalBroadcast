```python
import numpy as np
from tqdm import tqdm
%matplotlib inline
from matplotlib import pyplot as plt
import pickle
```

## Variable number of processes

```python
# repetitions
repetitions = range(5)

# n processes
ns = list(range(2, 7))

# messages
m = 10000
```

```python
results = pickle.loads(open('results_n_lcb.pkl', 'rb').read())
```

```python
xvals = ns
yvals_means = np.array([np.mean(results[n]) for n in ns])
yvals_stds  = [np.std(results[n]) for n in ns]
```

```python
# fitting a polynomial with deg 2
p = np.polyfit(xvals, yvals_means, deg = 2)
poly_vals = [p[0] * x ** 2 + p[1] * x + p[2] for x in xvals]
```

```python
plt.figure()
plt.xlabel('n')
plt.ylabel('Time, sec')
plt.title('URB for m=%d and different n' % m)
plt.plot(xvals, yvals_means, label = 'mean')
plt.fill_between(xvals, yvals_means - yvals_stds, yvals_means + yvals_stds, alpha = 0.5, label = 'std')
plt.plot(xvals, poly_vals, label = 'Deg 2 regression')
plt.legend()
#plt.yscale('log')
plt.show()
```
